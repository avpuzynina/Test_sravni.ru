```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

tables = pd.read_html('table_birth.html', header=0)
print('Tables found:', len(tables))
```

```python
df1 = tables[0]
print(df1)
```

```python
xticks = df1.iloc[:,0]
try:
    bars2 = df1.iloc[:,1].str.replace(' ','').astype('float')
except AttributeError:
    bars2 = df1.iloc[:,1].astype('float')
```

```python
barWidth1 = 0.065
barWidth2 = 0.065
x_range = np.arange(len(bars2) / 8, step=0.125)

plt.bar(x_range, bars2, color='#ffc001', width=barWidth2, edgecolor='#c3d5e8', label='Количество родившихся чел.')
# for i, bar in enumerate(bars2):
#      plt.text(i / 8 - 0.015, bar + 1, bar, fontsize=14)

plt.xticks(x_range, xticks)
plt.tick_params(
    bottom=False,
    left=False,
    labelsize=8
)
plt.rcParams['figure.figsize'] = [35, 15]
plt.axhline(y=0, color='gray')
plt.tick_params(axis='y', which='major', labelsize=25)
plt.legend(frameon=False, loc='lower center', bbox_to_anchor=(0.25, -0.1, 0.5, 0.5), prop={'size':20})
# plt.box(False)
# plt.savefig('plt', bbox_inches = "tight")
plt.show()
```
